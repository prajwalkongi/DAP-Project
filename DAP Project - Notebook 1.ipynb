### Reading XML File

```python
def read_xml(xml_file):
    try:
        tree = ET.parse(xml_file)
        root = tree.getroot()
        print('Succesfuly parsed the XML!')
        return root
        
    except FileNotFoundError:
        print('Error: Please check if the file exists')
        return None
    
    except ET.ParseError:
        print('Error: Failed to parse the XML file')
        return None
    
    except OSError:
        print('OSError: Error reading file: failed to load external entity')
        return None
```

```python
path = 'C:\\Users\\PRAJWAL\\Desktop\\NCI\\Projects\\DAP\\Dataset'

root = read_xml(path + '\\Statewide_Greenhouse_Gas_Emissions__Beginning_1990.xml')

# Statewide_Greenhouse_Gas_Emissions__Beginning_1990.xml
```

```python
# creating a list with the column names, we'll use this list to append our data

greenhouse_data = [['gross', 'net', 'conventional_accounting', 'economic_sector', 'sector',
                      'category', 'sub_category_1', 'sub_category_2', 'sub_category_3', 'year',
                      'gas', 'mt_co2e_ar5_20_yr', 'mt_co2e_ar4_100_yr']]
```

```python
for dat in root.findall('.//row'):
    
    gross = dat.find('.//gross').text
    net = dat.find('.//net').text
    conventional_accounting = dat.find('.//conventional_accounting').text
    economic_sector = dat.find('.//economic_sector').text
    sector = dat.find('.//sector').text
    category = dat.find('.//category').text
    sub_category_1 = dat.find('.//sub_category_1').text
    sub_category_2 = dat.find('.//sub_category_2').text
    sub_category_3 = dat.find('.//sub_category_3').text
    year = dat.find('.//year').text
    gas = dat.find('.//gas').text
    mt_co2e_ar5_20_yr = dat.find('.//mt_co2e_ar5_20_yr').text
    mt_co2e_ar4_100_yr = dat.find('.//mt_co2e_ar4_100_yr').text

    print(gross, net, conventional_accounting, economic_sector, sector, category, sub_category_1, 
          sub_category_2, sub_category_3, year, gas, mt_co2e_ar5_20_yr, mt_co2e_ar4_100_yr)

    data = [gross, net, conventional_accounting, economic_sector, sector, category, sub_category_1, 
                      sub_category_2, sub_category_3, year, gas, mt_co2e_ar5_20_yr, mt_co2e_ar4_100_yr]

    greenhouse_data.append(data)
```

```python
df1 = pd.DataFrame(greenhouse_data, columns = greenhouse_data[0]).iloc[1:]

df1.head()
```
